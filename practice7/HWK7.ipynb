```python

# Create RA column in decimal degrees
df['RA'] = (df['RAh'] * 15) + (df['RAm'] / 4) + (df['RAs'] / 240)

# Create Dec column in decimal degrees
df['Dec'] = df['DEd'] + (df['DEm'] / 60) + (df['DEs'] / 3600)
df.loc[df['DE-'] == '-', 'Dec'] *= -1  # Apply negative sign only to southern declinations

df[['RA', 'Dec']]
```

```python
df['SpType'] = df['SpType'].astype(str)

def spectral_type_to_color(spectral_type):
    if spectral_type.startswith('O'):
        return 'violet'  # or 'deep blue'
    elif spectral_type.startswith('B'):
        return 'medium blue'
    elif spectral_type.startswith('A'):
        return 'light blue'
    elif spectral_type.startswith('F'):
        return 'green'
    elif spectral_type.startswith('G'):
        return 'yellow'
    elif spectral_type.startswith('K'):
        return 'orange'
    elif spectral_type.startswith('M'):
        return 'red'
    else:
        return 'black'  

df['Color'] = df['SpType'].apply(spectral_type_to_color)

df[['SpType', 'Color']]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def adjust_ra(ra):
    ra_deg = ra * 15  # Convert hours to degrees
    return np.where(ra_deg > 180, ra_deg - 360, ra_deg)

def adjust_glon(glon):
    return np.where(glon > 180, glon - 360, glon)

def correct_dec(dec):
    # Check if declination is in degrees or hours
    if np.all(dec >= 0) and np.max(dec) <= 24:
        # If in hours, convert to degrees
        dec = dec * 15
    
    # Ensure declination is within -90 to 90 range
    dec = np.clip(dec, -90, 90)
    return dec

# Get GLON and GLAT
galactic_longitude = adjust_glon(df['GLON'].dropna().values)
galactic_latitude = df['GLAT'].dropna().values

# Get RA and Dec
equatorial_longitude = adjust_ra(df['RA'].dropna().values)
equatorial_latitude = correct_dec(df['Dec'].dropna().values)

plt.figure(figsize=(20, 10))

# Plot Equatorial coordinates
ax1 = plt.subplot(121, projection='aitoff')
ax1.set_title('All-Sky Map in Equatorial Coordinates (Epoch 2000.0)')
ax1.grid(True, alpha=0.3)
sc1 = ax1.scatter(np.radians(equatorial_longitude), np.radians(equatorial_latitude), 
                  s=2, alpha=0.5, c='red', edgecolors='none')
ax1.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])

# Plot Galactic coordinates
ax2 = plt.subplot(122, projection='aitoff')
ax2.set_title('All-Sky Map in Galactic Coordinates')
ax2.grid(True, alpha=0.3)
sc2 = ax2.scatter(np.radians(galactic_longitude), np.radians(galactic_latitude), 
                  s=2, alpha=0.5, c='blue', edgecolors='none')

plt.tight_layout()
plt.savefig('corrected_sky_map.pdf', dpi=300, bbox_inches='tight')
plt.show()

```
