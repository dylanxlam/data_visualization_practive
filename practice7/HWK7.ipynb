### **Create two columns ra and dec in the data frame having the right ascension in decimal degrees and declination in decimal degrees.**


```python

# Create RA column in decimal degrees
df['RA'] = (df['RAh'] * 15) + (df['RAm'] / 4) + (df['RAs'] / 240)

# Create Dec column in decimal degrees
df['Dec'] = df['DEd'] + (df['DEm'] / 60) + (df['DEs'] / 3600)
df.loc[df['DE-'] == '-', 'Dec'] *= -1  # Apply negative sign only to southern declinations

df[['RA', 'Dec']]
```

### **Assign colors to the stars according to spectral types:**

```python
df['SpType'] = df['SpType'].astype(str)

def spectral_type_to_color(spectral_type):
    if spectral_type.startswith('O'):
        return 'violet'  # or 'deep blue'
    elif spectral_type.startswith('B'):
        return 'medium blue'
    elif spectral_type.startswith('A'):
        return 'light blue'
    elif spectral_type.startswith('F'):
        return 'green'
    elif spectral_type.startswith('G'):
        return 'yellow'
    elif spectral_type.startswith('K'):
        return 'orange'
    elif spectral_type.startswith('M'):
        return 'red'
    else:
        return 'black'  

df['Color'] = df['SpType'].apply(spectral_type_to_color)

df[['SpType', 'Color']]
```

### **Create an all-sky map of the stars in the Yale Bright Star Catalog under Aitoff projection**

```python

color_map = {
    'violet': ('O Spectra', 'darkviolet'),
    'medium blue': ('B Spectra', 'blue'),
    'light blue': ('A Spectra', 'skyblue'),
    'green': ('F Spectra', 'green'),
    'yellow': ('G Spectra', 'gold'),
    'orange': ('K Spectra', 'orange'),
    'red': ('M Spectra', 'red'),
    'black': ('Other', 'black')
}

def adjust_ra(ra):
    ra_deg = ra * 15  
    return np.where(ra_deg > 180, ra_deg - 360, ra_deg)

def adjust_glon(glon):
    return np.where(glon > 180, glon - 360, glon)

def correct_dec(dec):
    # Check if declination is in degrees
    if np.all(dec >= 0) and np.max(dec) <= 24:
        dec = dec * 15
    
    dec = np.clip(dec, -90, 90)
    return dec

# coordinates
galactic_longitude = adjust_glon(df['GLON'].dropna().values)
galactic_latitude = df['GLAT'].dropna().values
equatorial_longitude = adjust_ra(df['RA'].dropna().values)
equatorial_latitude = correct_dec(df['Dec'].dropna().values)

equatorial_mask = df['RA'].notna() & df['Dec'].notna()
galactic_mask = df['GLON'].notna() & df['GLAT'].notna()
equatorial_colors = df.loc[equatorial_mask, 'Color'].map(dict([(k, v[1]) for k, v in color_map.items()])).values
galactic_colors = df.loc[galactic_mask, 'Color'].map(dict([(k, v[1]) for k, v in color_map.items()])).values

plt.figure(figsize=(20, 10))

# equatorial coordinates
ax1 = plt.subplot(121, projection='aitoff')
ax1.set_title('All-Sky Map in Equatorial Coordinates (Epoch 2000.0)')
ax1.grid(True, alpha=0.3)
sc1 = ax1.scatter(np.radians(equatorial_longitude), np.radians(equatorial_latitude), 
                  s=2, alpha=0.5, c=equatorial_colors, edgecolors='none')

#galactic coordinates
ax2 = plt.subplot(122, projection='aitoff')
ax2.set_title('All-Sky Map in Galactic Coordinates')
ax2.grid(True, alpha=0.3)
sc2 = ax2.scatter(np.radians(galactic_longitude), np.radians(galactic_latitude), 
                  s=2, alpha=0.5, c=galactic_colors, edgecolors='none')

# legend
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', 
                             markerfacecolor=color[1], label=color[0],
                             markersize=8) 
                  for color in color_map.values()]
ax1.legend(handles=legend_elements, loc='upper right', 
          bbox_to_anchor=(1.15, 1.0), title='Spectral Types')

plt.tight_layout()
plt.savefig('colored_sky_map.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
