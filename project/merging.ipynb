```python
import pandas as pd

# Read the CSV file from within the zip
mental_health = pd.read_csv('/Users/dylanlam/Documents/GitHub/data_visualization_practive/project/Mental health Depression disorder Data.csv.zip', compression='zip')
fat_supply = pd.read_csv('/Users/dylanlam/Documents/GitHub/data_visualization_practive/project/Fat_Supply_Quantity_Data.csv')
protein_supply = pd.read_csv('/Users/dylanlam/Documents/GitHub/data_visualization_practive/project/Protein_Supply_Quantity_Data.csv')
food_supply = pd.read_csv('/Users/dylanlam/Documents/GitHub/data_visualization_practive/project/Food_Supply_kcal_Data.csv')
```

```python
import pandas as pd

# Merging mental_health (using 'entity') and fat_supply on 'country'
merged_df = pd.merge(mental_health, fat_supply, left_on='Entity', right_on='Country', how='inner')

# Merging the resulting DataFrame with protein_supply on 'country'
merged_df = pd.merge(merged_df, protein_supply, on='Country', how='inner')

# Merging the resulting DataFrame with food_supply on 'country'
merged_df = pd.merge(merged_df, food_supply, on='Country', how='inner')

```
