```python
import pandas as pd

# Merging mental_health (using 'entity') and fat_supply on 'country'
merged_df = pd.merge(mental_health, fat_supply, left_on='Entity', right_on='Country', how='inner')

# Merging the resulting DataFrame with protein_supply on 'country'
merged_df = pd.merge(merged_df, protein_supply, on='Country', how='inner')

# Merging the resulting DataFrame with food_supply on 'country'
merged_df = pd.merge(merged_df, food_supply, on='Country', how='inner')

```

```python
import plotly.express as px

# Assuming your DataFrame is named 'merged_df'
fig = px.choropleth(subset, locations="Code",  # Replace with your actual country code column name
                    color="Schizophrenia (%)",  # Replace with your data column name
                    hover_name="Entity",  # Replace with your country name column name
                    hover_data=[
                        "Population",  # Add other statistics as needed
                        "Schizophrenia (%)"  # Add data column for hover display with custom formatting
                    ],
                    color_continuous_scale="Viridis",
                    title="Global Schizophrenia Rates"  # Add a title for the visualization
                    )

# Customize hover template for a cleaner display
fig.update_traces(hovertemplate="<b>%{hover_name}</b><br><br>" +
                   "%{Schizophrenia (%):.2f}% Schizophrenia<br>" +
                   "Population: %{Population:,}<br>" +
                   "GDP: %{GDP:.2f} (USD)")  # Replace 'USD' with your actual currency symbol

# Show all countries by setting a very low zoom level and centering on the equator
fig.update_layout(mapbox_style="carto-dark",  # Choose a map style
                   mapbox_zoom=0,  # Set a very low zoom level to show all countries
                   mapbox_center={"lat": 0, "lon": 0})  # Center on the equator

# Enable WebGL for potentially improved performance
fig.update_layout(config={"editable": True, "modeBarButtonsToAdd": ["drawcircle", "drawrect", "eraseshape"]})

fig.show()
```
