**Q. 6a (10 points)**.  
Which country in each of the continents - Africa, Asia, Europe, Latin America and  the Caribbean, Northern America, and Oceania - had the largest total immigration? Print the continent, country, and total immigration.

```python
country_continent_totals = df.groupby(['Country', 'Continent'])['Total'].sum().reset_index()

highest_immigration_per_continent = country_continent_totals.loc[country_continent_totals.groupby('Continent')['Total'].idxmax()]

for _, row in highest_immigration_per_continent.iterrows():
    print(f"Continent: {row['Continent']}, Country: {row['Country']}, Total Immigration: {row['Total']}")
```

**Q. 7a (10 points)**.  
Print the name of each country and the range (max - min) of the total immigration.

```python
years = [str(year) for year in range(1980, 2014)]

df['Immigration Range'] = df[years].max(axis=1) - df[years].min(axis=1)

for index, row in df.iterrows():
    print(f"Country: {index}, Range of Total Immigration: {row['Immigration Range']}")
```

**Q. 7b (10 points)**.  
Create boxplots showing the immigration of all countries over the years 1980 - 2013.
* Your graph should contain 134 boxplots, one boxplot per country, and each boxplot will display the immigration data from 1980 - 2013 for that country.
* You may need to use `pd.melt()` or `pd.transpose()` to arrange your data in the correct format for graphing.
* For graphing, we recommend using seaborn, but matplotlib is also acceptable. You should also create appropriate labels and modify them (if needed) to make them readable.

```python
df = df.reset_index()

df.rename(columns={df.columns[0]: 'Country'}, inplace=True)

graph_df = pd.melt(df, id_vars=['Country'], value_vars=years, var_name='Year', value_name='Immigration')

plt.figure(figsize=(20, 10))
sns.boxplot(x='Country', y='Immigration', data=graph_df)
plt.xticks(rotation=90)
plt.title('Immigration Boxplots for All Countries (1980 - 2013)')
plt.xlabel('Country')
plt.ylabel('Immigration Count')
plt.show()
```

**Q. 8 (20 points)**    
For the countries in the Region named as Caribbean, find the total immigration in the two decades (1980 to 1989) and (1990 - 1999). Print the decades as (Eighties and Nineties) and the total immigration for these decades from each of the Caribbean countries.

```python
caribbean = df[df['Region'] == 'Caribbean'].copy()

eighties = [str(year) for year in range(1980, 1990)]
nineties = [str(year) for year in range(1990, 2000)]

caribbean.loc[:, 'Eighties'] = caribbean[eighties].sum(axis=1)
caribbean.loc[:, 'Nineties'] = caribbean[nineties].sum(axis=1)

caribbean_immigration = caribbean[['Country', 'Eighties', 'Nineties']]

for index, row in caribbean.iterrows():
    print(f"Country: {row['Country']}, Total Immigration in the Eighties: {row['Eighties']}, Total Immigration in the Nineties: {row['Nineties']}")
```

**Q. 9 (20 points)**.  
The Scandinavian countries are Denmark, Norway, and Sweden. For each of the years (from 1980 to 2013), find the difference between the country with the highest immigration and the country with the lowest immigration. Print the year and the difference.

```python

scandinavian_df = df[df['Country'].isin(['Denmark', 'Norway', 'Sweden'])]


for year in years:
    max = scandinavian_df[year].max()
    min = scandinavian_df[year].min()
    
    difference = max - min
    
    print(f"Year: {year}, Difference: {difference}")
```

**Q. 10 (20 points)**.  
Find the countries in Africa that had the highest and lowest immigration for each of the years. Print the year, name of the country with highest immigration, immigration number, name of the country with the lowest immigration, and the immigration number. 

```python
african_df = df[df['Continent'] == 'Africa']

for year in years:
    max_immigration_country = african_df.loc[african_df[year].idxmax()]['Country']
    country_max = african_df[year].max()
    
    min_immigration_country = african_df.loc[african_df[year].idxmin()]['Country']
    country_min = african_df[year].min()
    
    print(f"Year: {year}, Highest: {max_immigration_country} ({country_max}), Lowest: {min_immigration_country} ({country_min})")
```
